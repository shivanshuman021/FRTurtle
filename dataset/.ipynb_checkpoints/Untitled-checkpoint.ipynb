```python
train = pd.read_csv('database.csv')
```

```python
train = train.iloc[:,1:3]
```

```python
train
```

```python
database = {}
for i in range(len(train)):
	k = train.iloc[i,1].split()
	k = k[1:]
	k[-1] = k[-1][:2]
	database[train.iloc[i,0]] = np.array(k,dtype='float64')
```
