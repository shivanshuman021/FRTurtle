```python
train = pd.read_csv('database.csv')
```

```python
train = train.iloc[:,1:3]
```

```python
database = {}
for i in range(len(train)):
	k = train.iloc[i,1].split()
	k = k[1:]
	x = k[-1]
	x = x[:-2]
	k[-1] = float(x)
	database[train.iloc[i,0]] = np.array(k,dtype='float64')
```
